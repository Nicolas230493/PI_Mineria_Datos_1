## Respuesta a las preguntas de análisis

### 1. ¿El plan de suscripción cambia según el país o la edad?
El análisis bivariado mostró que la distribución de planes es similar en todos los países, sin que ninguno muestre una preferencia marcada por un plan en particular. Respecto a la edad, tampoco se observó una correlación clara: usuarios de distintos rangos etarios se distribuyen de forma homogénea entre los tres planes. El plan Básico domina en todos los segmentos con aproximadamente el 45% de los usuarios.

### 2. ¿Los usuarios que más consumen son los que abren más tickets?
No. El heatmap de correlación mostró una correlación de -0.01 entre minutos de consumo mensual y tickets de soporte, prácticamente nula. Consumir más contenido no implica abrir más tickets de soporte técnico.

### 3. ¿Hay géneros que predominen según el rango de edad?
El análisis multivariado no mostró diferencias significativas en la preferencia de géneros según la edad. Las medianas de edad se mantienen estables entre los 30 y 40 años para todos los géneros, sin que ningún género esté asociado a un rango etario particular.

### 4. ¿Los nulos en tiempo de visualización se concentran en algún plan?
Esta pregunta fue respondida indirectamente durante la limpieza: los nulos en monthly_watch_time_mins se distribuyeron de manera uniforme entre los planes de suscripción, sin concentración en ninguno en particular.

## Hallazgos generales

- El dataset representa una base de usuarios latinoamericanos distribuida de forma equilibrada entre 7 países y 3 planes de suscripción.
- Las variables numéricas (edad, minutos de consumo, tickets de soporte) no presentan correlaciones significativas entre sí.
- El perfil típico del usuario es adulto joven (mediana de 33 años), con un consumo mensual moderado de alrededor de 750 minutos y una interacción mínima con el soporte técnico.
- El PCA reveló que el dataset no tiene variables fuertemente correlacionadas, lo que limita la capacidad de reducción de dimensionalidad: se necesitan 13 componentes para explicar el 80% de la varianza.
