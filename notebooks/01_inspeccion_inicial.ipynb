```python
# Analizar las fechas rotas sin alterar el dataset original
print("--- ANÁLISIS DE CALIDAD EN FECHAS ---")

# Trabajamos sobre una copia temporal de la columna sin nulos para no romper nada
columna_fecha = df['last_login_date'].dropna()

# 1. Detectar textos basura que no son fechas
textos_basura = columna_fecha[columna_fecha.isin(['not_available', '0000-00-00'])]
print(f"Registros con texto inválido ('not_available'): {len(textos_basura)}")

# 2. Detectar fechas con problemas de lógica (meses mayores a 12 o días imposibles)
# Usamos to_datetime de forma aislada para ver qué falla, sin tocar el dataframe
fechas_limpias_temp = columna_fecha[~columna_fecha.isin(['not_available', '0000-00-00'])]
imposibles = fechas_limpias_temp[pd.to_datetime(fechas_limpias_temp, errors='coerce', format='%Y-%m-%d').isna()]
print(f"Registros con fechas lógicamente imposibles (ej. mes 15 o día 40): {len(imposibles)}")
if len(imposibles) > 0:
    print(f"Ejemplos de fechas imposibles: {imposibles.head(5).values}")

# 3. Detectar fechas que viajan al futuro (posteriores a hoy)
# Primero convertimos temporalmente lo que sí se puede para comparar
fechas_validas_convertidas = pd.to_datetime(fechas_limpias_temp, errors='coerce', format='%Y-%m-%d')
futuras = fechas_validas_convertidas[fechas_validas_convertidas > '2026-06-27']
print(f"\nRegistros con fechas en el futuro (posteriores a junio 2026): {len(futuras)}")
if len(futuras) > 0:
    # Mostramos cómo se ven originalmente en el formato sucio
    print(f"Ejemplos de fechas futuras detectadas: {df.loc[futuras.index, 'last_login_date'].head(5).values}")
```

### Lo que encontré al revisar los datos en bruto

El dataset llegó con varios problemas de calidad que voy a tener que corregir antes de cualquier análisis. Tiene 8160 filas y 8 columnas, mezclando datos personales con métricas de consumo.

**Valores faltantes:** No son una locura, pero hay que atenderlos. Me faltan 193 tiempos de visualización, 240 géneros favoritos y 320 fechas de último login. El resto está completo.

**Duplicados:** Hay 126 filas calcadas que ensucian los números. Pero el problema más feo son los 160 `user_id` repetidos (80 usuarios con dos registros). Algunos tienen datos distintos, así que no sé si son actualizaciones mal cargadas o errores de base. Voy a tener que definir un criterio para decidir con cuál quedarme.

**Categorías inconsistentes:** Un quilombo. `subscription_plan` tiene 15 variantes para 3 planes que debería tener. `country` aparece de 26 formas distintas (Brasil, Brazil, brasil, BRA...). Y `favorite_genre` mezcla inglés, español, mayúsculas y errores ortográficos. Todo va a tener que unificarse.

**Outliers:** Hay cosas que no cierran. Edades negativas y gente de más de 100 años. Tiempos de visualización en negativo y otros que superan los 10.000 minutos (imposible en un mes). Tickets de soporte negativos. Todo eso hay que revisarlo.

**Fechas rotas:** `last_login_date` es texto, no fecha. Tiene `not_available`, `0000-00-00`, fechas en 2029, meses 15 y formatos mezclados (YYYY-MM-DD, DD-MM-YYYY, etc.). Un dolor de cabeza que voy a tener que normalizar.

**El plan:** Limpiar duplicados, unificar categorías, corregir outliers, arreglar las fechas y decidir qué hacer con los nulos. Recién después de eso voy a poder analizar en serio.
